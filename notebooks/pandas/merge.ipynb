```python
import pandas as pd
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2
```

```python
df = pd.merge(df1,df2,on="city")
df
```

```python
df1.join(df2,lsuffix="_l",rsuffix="_r")
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35, 38],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,71],
})
df2
```

```python
df=pd.merge(df1,df2,on="city")
df
```

```python
df=pd.merge(df1,df2,on="city",how="outer")
df
```

```python
df=pd.merge(df1,df2,on="city",how="left")
df
```

```python
df=pd.merge(df1,df2,on="city",how="right")
df
```

```python
df=pd.merge(df1,df2,on="city",how="outer",indicator=True)
df
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
df2
```

```python
df=pd.merge(df1,df2,on="city",how="outer",suffixes=("-first","-second"))
df
```

```python
df=pd.merge(df1,df2,on="city",how="outer",suffixes=("-uno","-duno"))
df
```
