```python
import pandas as pd
india_weather={
    "city":["mumbai","dehli","banglore"],
    "temperature":[32,45,22],
    "humidity":[33,55,22]}
df1=pd.DataFrame(india_weather)
df1
```

```python
us_weather={
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]}
df2=pd.DataFrame(us_weather)
df2
```

```python
df=pd.concat([df1,df2],ignore_index=True)
df
```

```python
df=pd.concat([df1,df2],keys=["ind","us"])
df
```

```python
df.loc["ind"]
```

```python
df=pd.concat([df1,df2],axis=1)
df
```

```python
temp_weather={
    "city":["mumbai","dehli","banglore"],
    "temperature":[32,45,22],
    }
df3=pd.DataFrame(temp_weather,index=[0,1,2])
df3
```

```python
df=pd.concat([df3,df4],axis=1)
df
```
