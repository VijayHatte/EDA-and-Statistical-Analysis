```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```

```python
plt.pie(exp_vals,labels=exp_labels)
```

```python
plt.pie(exp_vals,labels=exp_labels,shadow=True)
plt.axis("equal")
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels,radius=1,autopct='%1.1f%%')
plt.axis("equal")
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels,radius=1,autopct="%1.1f%%",explode=[0,0.2,0,0,0.1])
plt.axis("equal")
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels,radius=1,autopct="%1.1f%%",counterclock=False)
plt.axis("equal")
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels,radius=1,autopct="%1.1f%%",counterclock=True,startangle=90)
plt.axis("equal")
plt.show()
```
