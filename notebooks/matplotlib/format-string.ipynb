```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.title("weather")
plt.xlabel("days")
plt.ylabel("temperature")
plt.plot(x,y,'g--s',markersize=20,alpha=1)
```

```python
plt.title("weather")
plt.xlabel("days")
plt.ylabel("temperature")
plt.plot(x,y,'b-.2',markersize=20,alpha=1)
```

```python
plt.title("weather")
plt.xlabel("days")
plt.ylabel("temperature")
plt.plot(x,y,'--8',color="#085544",markersize=20,alpha=1)
```
