```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
xpos=np.arange(len(company))
xpos
```

```python
plt.title("US Tech Stocks")
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.bar(xpos,revenue,label="revenue")
plt.xticks(xpos,company)
plt.legend()
```

```python
profit=[40,2,34,12]
```

```python
plt.title("US Tech Stocks")
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.bar(xpos-0.2,revenue,width=0.4,edgecolor="black",label="revenue")
plt.bar(xpos+0.2,profit,width=0.4,edgecolor="black",label="profit")
plt.xticks(xpos,company)
plt.legend()
plt.savefig(r"C:\Users\ASUS\Downloads\vij.pdf",bbox_inches="tight",pad_inches=1,transparent=True)
```

```python
plt.title("US Tech Stocks")
plt.xlabel("Revenue")
plt.ylabel("Company")
plt.barh(xpos-0.2,revenue,label="revenue")
plt.barh(xpos+0.2,profit,label="profit")
plt.xticks(xpos,company)
plt.legend()
```
