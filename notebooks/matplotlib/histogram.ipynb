```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
```

```python
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.8)
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
```

```python
plt.xlabel("Sugar range")
plt.ylabel("No.of patients")
plt.title("Sugar level analysis")
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.8,color=['y','g'],label=['Men','Women'])
plt.legend()
```

```python
plt.xlabel("Sugar range")
plt.ylabel("No.of patients")
plt.title("Sugar level analysis")
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.8,color=['y','g'],label=['Men','Women'],histtype="step")
plt.legend()
```

```python
plt.xlabel("Sugar range")
plt.ylabel("No.of patients")
plt.title("Sugar level analysis")
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.8,color=['y','g'],label=['Men','Women'],orientation="horizontal")
plt.legend()
```
